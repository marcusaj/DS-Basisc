```python
import numpy as np
import pandas as pd
```

Use dataset about Covid World Vaccination Progress: https://www.kaggle.com/gpreda/covid-world-vaccination-progress

```python
vacc = pd.read_csv('country_vaccinations.csv')
```

```python
vacc
```

```python
#General info about rows and columns
vacc.info()
```

```python
#Min for each column
vacc.min()
```

```python
#Max number of vaccines inoculated per day within all countries from dataset. 
#The number looks too high. Most probably it is a typo.
print("max daily vaccines innoculated:", vacc['daily_vaccinations'].max())
```

```python
#Finding out the country and the date for max daily vaccines innoculated
print("Country with max daily vaccines inoculated:", vacc.loc[vacc['daily_vaccinations'].argmax(), 'country'])
print("Date:", vacc.loc[vacc['daily_vaccinations'].argmax(), 'date'])
```

```python
#Coutnry with the highest number of people vaccinated
print("Country with highest number of people vaccinated on", vacc.loc[vacc['people_vaccinated'].argmax(), 'date'], ":", vacc.loc[vacc['people_vaccinated'].argmax(), 'country']) 
print( "Total people vaccinated  until that date (inclusive):", vacc['people_vaccinated'].max())
```

```python
#Coutnry with the highest share of people fully vaccinated
print("Country with highest number of people vaccinated on", vacc.loc[vacc['people_fully_vaccinated_per_hundred'].argmax(), 'date'], ":", vacc.loc[vacc['people_fully_vaccinated_per_hundred'].argmax(), 'country']) 
print( "Share of people vaccinated  until that date (inclusive):", vacc['people_fully_vaccinated_per_hundred'].max())
print("the share looks like they vaccinated and their neighbouring countries :)")
```

```python
#Number of people vaccinated in Modolva
print("Number of people vaccinated in Modolva:",vacc.loc[vacc['country']== 'Moldova', 'people_vaccinated'].max())
```
