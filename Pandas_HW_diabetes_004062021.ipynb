```python
import numpy as np
import pandas as pd
```

```python
db = pd.read_csv('diabetes.csv')
```

```python
db
```

```python
#Average value per each attribute grouped by types of Pregnancies
db.groupby(['Pregnancies']).mean()
```

```python
#Delete one column,but first save it seprately to recover it back after
```

```python
#1) Saving column that will be deleted
DPF_column = db['DiabetesPedigreeFunction'].values
```

```python
#2) Delete column DiabetesPedigreeFunction
db = db.drop(['DiabetesPedigreeFunction'], axis = 1)
```

```python
#Showing that it's not in the db anymore
db
```

```python
#3) Recovering back column DiabetesPedigreeFunction
db['DiabetesPedigreeFunction'] = DPF_column
```

```python
#Showing that it's back again in the db, but as the last column now
db
```

```python
# Shuffle the dataset
db = db.sample(frac=1)
db
```

```python
#Create new column as a flag for normal glucose /abnormal glucose level  and mapping it 
db['NormalGlucoseLevel'] = db['Glucose'].apply(lambda x : x < 200   and x > 80 ).map({True: 1, False: 0})
```

```python
db
```

```python
#Calculating how many have normal glucose level and how many has abnormal level
db['NormalGlucoseLevel'].value_counts()
```
